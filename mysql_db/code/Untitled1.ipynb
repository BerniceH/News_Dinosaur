```python
import requests
import json
import time
import pymysql

#建立與mysql的連線
connd = pymysql.connect(host='db', port=3306, user='root', passwd='iii', db='chatbot_db',charset='utf8mb4')
curs = connd.cursor()

with open('news_emo_label_abstract_kw.json', 'r', encoding="utf-8") as f:
    all_news = json.load(f)
    
for a in all_news:
    
    if a['event'] == "honey_lemon":
        a['event'] = 'TLDR03'
    elif a['event'] == 'sleep':
        a['event'] = 'TLDR01'
    elif a['event'] == 'goldenhorse':
        a['event'] = 'TLDR04'
    elif a['event'] == 'pig':
        a['event'] = "TLDR02"
        
    curs.execute('SELECT title_id FROM chatbot_db.News WHERE url = ("%s")' % (a['url']))
    title_id = curs.fetchone()
    
    insertsql= "INSERT INTO chatbot_db.TLDR_News (tldr_id, title_id) VALUES (%s, %s)"
    value = (a['event'], title_id)
    curs.execute(insertsql , value)
    connd.commit()
    print("ok")
    time.sleep(1)

curs.close()
```

```python
import requests
import json
import time

ip_location='chatbot_api'

with open('news_emo_label_abstract_kw.json', 'r', encoding="utf-8") as f:
    all_news = json.load(f)

    for a in all_news:
        if a['news_emo']== "pos":
            a['news_emo'] = "positive"
        elif a['news_emo'] == "neg":
            a['news_emo'] = "negative"

        if a['source'] == 'TVBS':
            a['source'] = 'TVBS新聞台'
        elif a['source'] == 'StormMedia':
            a['source'] = '風傳媒 THE STORM MEDIA'

        if a["author"] == None:
            a["author"]= []

        if a["kw"] == None:
            a["kw"]= []

        News = {  
            "source": a['source'],
            "label": a['news_label'],
            "ariticle_emotion" : a['news_emo'],
            "title" : a['title'],
            "url" : a['url'],
            "date_" : a["date_"],
            "content" : a['content'],
            "img_url" : a['img_url'],
            "kw": a['kw'],
            "author": a['author'],
            "abstract": a['abstract'] 
        }

        # 將json傳回API Server
        Endpoint='http://%s:5000/news' % (ip_location)

        # header要特別註明是json格式
        Header={'Content-Type':'application/json'}

        # 傳送post對API server新增資料 
        Response=requests.post(Endpoint,headers=Header,data=json.dumps(News))
        time.sleep(1)

        #印出Response的資料訊息
        print(Response)
```
